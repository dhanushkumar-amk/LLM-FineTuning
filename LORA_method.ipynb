### **Import Everything**

```python
from unsloth import FastLanguageModel
from unsloth.chat_templates import get_chat_template
from trl import SFTTrainer, SFTConfig
from datasets import load_dataset
import torch
print("Imports done")
```

### **Load Model**

```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "HuggingFaceTB/SmolLM2-135M-Instruct",
    max_seq_length = 512,
    load_in_4bit = False,
)
print("Model loaded")
```

### **Add LoRA**


```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 8,
    lora_alpha = 16,
    lora_dropout = 0.05,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj"],
    use_dora = False,
)
model.print_trainable_parameters()
```

### **Load and Format Dataset**

```python
raw_dataset = load_dataset("tatsu-lab/alpaca", split="train[:500]")

def format_example(example):
    if example["input"] and example["input"].strip():
        user_message = example["instruction"] + "\n\n" + example["input"]
    else:
        user_message = example["instruction"]

    return {
        "text": (
            "<|im_start|>system\nYou are a helpful assistant.<|im_end|>\n"
            f"<|im_start|>user\n{user_message}<|im_end|>\n"
            f"<|im_start|>assistant\n{example['output']}<|im_end|>"
        )
    }

dataset    = raw_dataset.map(format_example, remove_columns=raw_dataset.column_names)
split      = dataset.train_test_split(test_size=0.1, seed=42)
train_data = split["train"]
test_data  = split["test"]

print("Train:", len(train_data))
print("Test:", len(test_data))
print("\nSample:")
print(train_data[0]["text"])
```

### **Train**

```python
trainer = SFTTrainer(
    model = model,
    train_dataset = train_data,
    eval_dataset = test_data,
    args = SFTConfig(
        output_dir = "./lora-smollm2",
        num_train_epochs = 2,
        per_device_train_batch_size = 4,
        gradient_accumulation_steps = 2,
        learning_rate = 2e-4,
        warmup_ratio = 0.03,
        lr_scheduler_type = "cosine",
        logging_steps = 10,
        eval_strategy = "steps",
        eval_steps = 50,
        fp16 = True,
        report_to = "none",
        seed = 42,
    ),
)

print("Starting training...")
trainer.train()
print("Training complete!")
```

### **Save**

```python
model.save_pretrained("./lora-adapter")
tokenizer.save_pretrained("./lora-adapter")
print("Saved")
```

### **Test**

```python
FastLanguageModel.for_inference(model)

messages = [
    {"role": "system",  "content": "You are a helpful assistant."},
    {"role": "user",    "content": "Give three tips for staying healthy."}
]

inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,
    add_generation_prompt = True,
    return_tensors = "pt"
).to("cuda")

outputs = model.generate(
    input_ids = inputs,
    max_new_tokens = 150,
    temperature = 0.7,
    do_sample = True,
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
